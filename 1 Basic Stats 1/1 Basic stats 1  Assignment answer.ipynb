# Bar Chart Analysis for Categorical Column:

```python
# Creating the bar chart
categories=df1['Day']
values=df1['Net Sales Value']
plt.figure(figsize=(10,7))
plt.bar(categories, values)

# Adding title and labels
plt.title('Bar Chart Example')
plt.xlabel('Day')
plt.ylabel('Net Sales Value')

# Displaying the chart
plt.show()
```

```python
#in this bar chart we can observe that there is  not much impact on the sales value of weekends and weekdays since all are
#almost same sales value
```

```python
# Creating the bar chart
categories=df1['Brand']
values=df1['Net Sales Value']
plt.figure(figsize=(10,7))
plt.bar(categories, values)

# Adding title and labels
plt.title('Bar Chart Example')
plt.xlabel('Day')
plt.ylabel('Net Sales Value')

# Displaying the chart
plt.show()
```

```python
# here we can conclude that first 4 brands are contrubuting majoring in sales value and other 5 has less impact
```

```python
# Creating the bar chart
categories=df1['Brand']
values=df1['Discount Amount']
plt.figure(figsize=(10,7))
plt.bar(categories, values)

# Adding title and labels
plt.title('Bar Chart Example')
plt.xlabel('Discount Amount')
plt.ylabel('Net Sales Value')

# Displaying the chart
plt.show()
```

```python
#here we can conclude that the first four brands which contribute majorly in sales value those brand giving  mojor discounts to
#customers    or this brand are giving more discounts that why customers are buying more from this brand's products
```

# Standardization of Numerical Variables

```python
#! pip install --upgrade seaborn --user
import seaborn as sn
```

```python
df=pd.read_csv('heights.csv')
```

```python
df.head()
```

```python
sn.histplot(df['height'], kde=True)
```

```python
df['Z-Score']=(df.height-df.height.mean())/df.height.std()
```

```python
df['Z-Score']
df.head()
```

```python
sn.histplot(df['Z-Score'],kde=True)
```

```python
#in first histplot data is ploted as it is and in second histplot   is plotted  with transformed data of z-score
```
