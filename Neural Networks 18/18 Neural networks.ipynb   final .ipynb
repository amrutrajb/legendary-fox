```python
#Model Implementation
#● Construct a basic ANN model using your chosen high-level neural network library.
#Ensure your model includes at least one hidden layer.
#● Divide the dataset into training and test sets.
#● Train your model on the training set and then use it to make predictions on the test set.
```

```python
#separating the data and target from the dataset
x=df_scaler.drop(columns=['letter'])
y=df_scaler["letter"]
```

```python
#creating training and test dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state=42)
```

```python
# create instance of Sequential class
model = Sequential()
# create neurons in input, hidden and output layers and assign random weights to input
# In I/P layer 13 neurons will be there as there are 13 independent features

#Add hidden layer, randomly 10 neurons are added
model.add(Dense(units=10,activation='relu'))# as target is binary use relu activation

model.add(Dense(units=20,activation='relu'))
model.add(Dense(units=10,activation='relu'))
#Add output layer
model.add(Dense(units=26, activation='softmax'))# for binary classification use sigmoid activation in output layer

#Establish the connection between the layers
model.compile(optimizer = 'adadelta',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
# adadelta is gradient descent algorithm for weight updation which we have selected randomly. We will hypertune it afterwards
# binary_crossentropy: also called log_loss, finds erros, range is 0 to 1

#Fit the data, perform forward and back propagation
model.fit(x_train,y_train, epochs=50)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# Define the model-building function
def create_model(units_hidden1=10, units_hidden2=80, units_hidden3=10, optimizer='adam'):
    model = Sequential()
    model.add(Dense(units=16, activation='relu', input_dim=16))  # Input layer with 13 neurons
    model.add(Dense(units=units_hidden1, activation='relu'))  # First hidden layer
    model.add(Dense(units=units_hidden2, activation='relu'))  # Second hidden layer
    model.add(Dense(units=units_hidden3, activation='relu'))  # Third hidden layer
    model.add(Dense(units=26, activation='softmax'))  # Output layer
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Wrap the model for compatibility with GridSearchCV
model = KerasClassifier(model=create_model, verbose=0)

# Define the grid of hyperparameters to tune
param_grid = {
    'model__units_hidden1': [8, 10, 12],   # Number of neurons in the first hidden layer
    'model__units_hidden2': [64, 68,72], # Number of neurons in the second hidden layer
    'model__units_hidden3': [8, 10, 12],   # Number of neurons in the third hidden layer
    'model__optimizer': ['adam', 'adadelta', ], # Optimizer choices
    'batch_size': [16, 32,],            # Batch size
    'epochs': [20, 50]                     # Number of epochs
}
```

```python
# Create the GridSearchCV object
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
```

```python
# Fit the grid search to the data
grid_result = grid.fit(x_train, y_train)
```

```python
# Display the best parameters and score
print(f"Best parameters: {grid_result.best_params_}")
print(f"Best score: {grid_result.best_score_}")
```
